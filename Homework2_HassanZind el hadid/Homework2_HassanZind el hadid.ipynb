```python
school_cdf = thinkstats2.Cdf(school_agemomb1_f, label='School')
university_cdf = thinkstats2.Cdf(university_agemomb1_f, label='University')

thinkplot.PrePlot(2)
thinkplot.Cdfs([school_cdf, university_cdf])
thinkplot.Config(xlabel='Age Labels', ylabel='CDF')

# CDFs of "school" and "university" data frames with respect to their columns "agemomb1"
```

```python
age_labels = range(0, 6)
diffs = []
for age_label in age_labels:
    p1 = school_pmf.Prob(age_label)
    p2 = university_pmf.Prob(age_label)
    diff = 100 * (p1 - p2)
    diffs.append(diff)
    
thinkplot.Bar(age_labels, diffs)
thinkplot.Config(xlabel = "Age_labels", ylabel = "Percentage Points", title = "Difference in PMFs")

# PMF of difference of age at first birth obtained from column "agemomb1" between "student" and "university" data frames.
# Note that upper bars represent "school" and lower bars represents "university".
```
